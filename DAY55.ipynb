```python
# Generating dataset from make regression
```

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
x,y=make_regression(n_samples=4,n_features=1,n_informative=1,n_targets=1,noise=80,random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(x,y)
```

```python
# applying OLS(ordinary least square)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=LinearRegression()
reg.fit(x,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
# Drawing the regression line
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red')
```

```python
# applying Gradient Descent algorithm assuming slope(m)=78.35
# and let the value of intercept(b)=0
y_pred=((78.35*x)+0).reshape(4)  # y=mx+b
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
# the above blue line is pointing at b=0 and through changing the b value
# we have to reach the red line
```

```python
m=78.35
b=0
```

```python
# Applying GD
# Loss func L= -2 summation of i=1 to n (yi-mxi-b)
loss_slope= -2 * np.sum(y- m*x.ravel()-b)
loss_slope
```

```python
# Let learning rate = 0.1
lr=0.1

step_size=loss_slope*lr  # learning rate * slope is called stepsize
step_size
```

```python
# Calculating new intercept
b=b-step_size
b
```

```python
y_pred1=((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred1,color='green',label='b=updated')
plt.plot(x,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope= -2 * np.sum(y- m*x.ravel()-b)
loss_slope
```

```python

step_size=loss_slope*lr  # learning rate * slope is called stepsize
step_size
```

```python
# Calculating new intercept
b=b-step_size
b
```

```python
y_pred2=((78.35*x)+b).reshape(4)

plt.scatter(x,y)
plt.plot(x,reg.predict(x),color='red',label='OLS')
plt.plot(x,y_pred2,color='black',label='b=updated2')
plt.plot(x,y_pred1,color='green',label='b=updated')
plt.plot(x,y_pred,color='blue',label='b=0')
plt.legend()
plt.show()
```

```python
# in the above code you can observe that the pred2 line is almost touching the
# red line we are close to our best fit line in GD after 2 to 3 iteration we can
# get our best fit line
```

```python
# Staring the iteration using loop

b=-100
m=78.35
lr=0.1

epochs=10

for i in range(epochs):
  loss_slope= -2 * np.sum(y- m*x.ravel()-b)
  b=b-(lr*loss_slope)
  y_pred=m*x+b
  plt.plot(x,y_pred)
plt.scatter(x,y)
```
